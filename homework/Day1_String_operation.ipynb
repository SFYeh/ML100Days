## 讀入所需文件

```python
import pandas as pd
#TSV文件有別於CSV的文件是使用\ t作為分隔符，CSV則是使用，作為分隔符。
dataset = pd.read_csv(r'Restaurant_Reviews.tsv', sep='\t')
```

```python
all_review = dataset['Review'].values
```

### 計算有多少個句子是以 . 結尾

```python
# 簡潔版寫法 (沒有要置換字串內容用這樣寫比較有效率)
n=0

for sentence in all_review:
    if sentence.endswith('.'):  #沒有寫True/False預設if的條件就是True 
        n+=1
print('There are {} sentences end with .'.format(n))
```

### 將所有. 換成 ,

```python
# 用index表示迴圈，方便置換字串
for i in range(len(all_review)):
    all_review[i]=all_review[i].replace('.',',')
```

### 將所有sentence 中的第一個 the 置換成 The
根據題目/解答的設計， whether, there 也會被調換就是了 (冏

```python
a='12345'
't'.join((a[:3],a[3+1:]))
```

```python
# 確認這個句子有"the" (避免沒有而回傳-1)
# 用.find()找到sentence中的第一個 the 的index
# 用join 做拼湊

for sentence_number in range(len(all_review)):
    input_sentence = all_review[sentence_number]
    if 'the' in input_sentence:
        location = input_sentence.find('the')
        input_sentence = ''.join((input_sentence[:location],'T',input_sentence[location+1:])) #:locatio & location:會剛好跳過location的那個字母
        all_review[sentence_number] = input_sentence     
        
"""下面這種寫法要注意寫index
for i in range(len(all_review)):
    if 'the' in all_review[i]:
        location=all_review[i].find("the")
        all_review[i]=''.join((all_review[i][:location],'T',all_review[i][(location+1):]))      
"""
```

### 將偶數句子全部轉換為大寫，基數句子全部轉換為小寫

```python
for i in range(len(all_review)):
    if (i+1)%2==0:
        all_review[i]=all_review[i].upper()  
    else:
        all_review[i]=all_review[i].lower()   
```
